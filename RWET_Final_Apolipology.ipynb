```python
import annoy
import random
```

```python
text = open("apology_text.txt").read()
words = text.split()
```

```python
import spacy
nlp = spacy.load('en_core_web_md')
```

```python
regretting = nlp(text)
```

```python
from simpleneighbors import SimpleNeighbors
```

```python
lookup = SimpleNeighbors(300)
for sent in regretting.sents:
    lookup.add_one(sent.text.replace("\n", " "), sent.vector)
lookup.build()
```

```python
say_it = lookup.nearest(nlp("we won't forget").vector)
say_it
```

```python
chunk_lookup = SimpleNeighbors(300)
for chunk in regretting.noun_chunks:
    chunk_lookup.add_one(chunk.text.replace("\n", " "), chunk.vector)
chunk_lookup.build()
```

```python
consequences = chunk_lookup.nearest(nlp("responsibility").vector)
```

```python
def vec(s):
    return nlp.vocab[s].vector
```

```python
from collections import defaultdict
import numpy as np
```

```python
word_lookup = SimpleNeighbors(300)
for item in nlp.vocab:
    if item.has_vector and item.prob > -15 and item.is_lower:
        word_lookup.add_one(item.text, item.vector)
word_lookup.build()
```

```python
from itertools import islice
```

```python
qualified = [item for item in nlp.vocab if item.has_vector and item.is_alpha]
```

```python
lexmap = []
t = annoy.AnnoyIndex(300)
for i, item in enumerate(islice(sorted(qualified, key=lambda x: x.prob, reverse=True), 100000)):
    t.add_item(i, item.vector)
    lexmap.append(item)
t.build(25)
```

```python
def similarsemantic(t, nlp, word, n):
    seen = set()
    count = 0
    for i in t.get_nns_by_vector(nlp.vocab[word].vector, 100):
        this_word = lexmap[i].orth_.lower()
        if this_word not in seen and word != this_word:
            seen.add(this_word)
            count += 1
            yield this_word
            if count >= n:
                break
remark = list(similarsemantic(t, nlp, "remark", 100))
remark
```

```python
remarkWords = ' '.join(word for word in remark)
remarkWords
```

```python
remarking = nlp(remarkWords)
```

```python
remarkAdj = [item.text for item in remarking if item.pos_ == 'ADJ']
remarkAdj
```

```python
shame = list(similarsemantic(t, nlp, "shame", 100))
shame
```

```python
shameWords = ' '.join(word for word in shame)
shameWords
```

```python
shaming = nlp(shameWords)
```

```python
shameNoun = [item.text for item in shaming if item.pos_ == 'NOUN']
shameNoun 
```
